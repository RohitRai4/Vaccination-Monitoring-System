```python
import datetime
from datetime import date

def vaccination():
    "This function show candidate vaccination status with Date"
    # Admin login portal
    attempts = 3
    while attempts > 0:
        print(f"Welcome to Vaccination Portal")
        admin_id = input("Enter your admin id: ")
        password = input("Enter your password: ")
    
        if admin_id == "rohit" and password == "xyz":
            print("Login successful!")
            break
        else:
            attempts -= 1
            print(f"Invalid username or password. {attempts} attempts remaining.")

            if attempts == 0:
                print("Login failed. Maximum attempts reached.")
        print()       
    # Candidate Portal Login            
    while True:    
        username = input("Enter your Candidate id : ")

        for name in df[df["userid"] == username]["name"].values:
            print(f"Candidate Name: ",name)
        for status1 in df[df["userid"] == username]["1st dose status"].values:
            print(f"First Dose Status: ",status1)
        for status2 in df[df["userid"] == username]["2nd dose status"].values :
            print(f"Second Dose Status: ",status2)
        

            # Show the Status of 1st and 2nd dose:
            if status1=="yes" and status2=="yes":
                print(f"Hi {name} ! congrulatins you have completed vaccinated")
            elif status1=="no" and status2=="no":
                print(f"Hi {name} ! Please take your first dose at earliest")
            elif status1=="yes" and status2=="no":

    
                d1_date =df[df["userid"]==username]["date of 1st dose"].values[0]
                d1 = datetime.datetime.strptime(d1_date, "%d-%m-%Y").date()

                print(f"your 1st Dose Date : {d1_date}")
        
               # now date
                d2 = date.today()
                delta = (d2 - d1)


                # Check the second dose overdue day:    
                if delta.days > 90:
                    a=delta.days
                    print(f"Hi {name} ! Please take your 2nd dose at the earliest: {a}  overdue.")

                # check the second dose date after first dose:    
                elif delta.days < 90:
                    d3 = 90 - delta.days
                    d4 = d2 + d3
                    print(f"Hi {name} ! Please take your 2nd dose on or after: {d4}" )
            print()

    
```
